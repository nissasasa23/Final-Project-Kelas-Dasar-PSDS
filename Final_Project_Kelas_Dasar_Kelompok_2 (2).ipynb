```python
# Menampilkan dataset produk olahan ikan yang ada di Wilayah Maluku - Papua
produk_olahan_ikan_Maluku_Papua = produk_olahan_ikan[produk_olahan_ikan.Wilayah == 'Maluku - Papua']
produk_olahan_ikan_Maluku_Papua
```

```python
# Melihat volume yang paling besar tersebut berada di Provinsi mana
produk_olahan_ikan_Maluku_Papua[produk_olahan_ikan_Maluku_Papua.Volume == 388217]
```

```python
# Visualisasi data
Nama_Provinsi = produk_olahan_ikan_Maluku_Papua.NamaProvinsi
Volume = produk_olahan_ikan_Maluku_Papua.Volume

plt.figure(figsize = (10,5))
plt.bar(Nama_Provinsi, Volume, color = 'thistle')

plt.title('Produk Olahan Ikan di Wilayah Maluku - Papua Tahun 2001-2011', size = 16)
plt.xlabel('Nama Provinsi', size = 14)
plt.ylabel('Volume', size = 14)
plt.xticks(size = 12)
plt.yticks(size = 12)

plt.show()
```

### Insight 4

**Kenapa pada Wilayah Maluku - Papua, Provinsi Maluku menghasilkan produk olahan ikan yang paling banyak?**

```python
# Visualisasi data produk olahan ikan berdasarkan nama provinsi
produk_olahan_ikan.groupby(['Tahun','NamaProvinsi'])['Volume'].sum().unstack().plot(marker='.', cmap='plasma')
plt.title('Volume Produk Olahan Ikan pada Tahun 2001-2011',loc='center',pad=30, fontsize=25, color='black')
plt.xlabel('Tahun', fontsize = 20)
plt.ylabel('Volume',fontsize = 20)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/100000).astype(int))
plt.legend(loc='upper center', bbox_to_anchor=(1.1, 1), shadow=True, ncol=1)
plt.gcf().set_size_inches(25,15)
plt.tight_layout()
plt.show()
```

Berdasarkan grafik diatas dari tahun 2001 sampai dengan 2012, Wilayah Maluku - Papua memperoleh jumlah volume produk olahan ikan tertinggi dibandingkan Wilayah lainnya.

```python
# Visualisasi data volume produk olahan ikan diwilayah Maluku - Papua berdasarkan jenis olahan
plotnine.options.figure_size = (10,10)
(ggplot(produk_olahan_ikan[produk_olahan_ikan['NamaProvinsi'].isin(['Maluku','Maluku Utara','Papua Barat','Papua'])])

+ aes(x = 'JenisOlahan', y = 'Volume', fill = 'NamaProvinsi')
+ geom_col(position = 'position_dodge')
+ coord_flip()
+ labs(title = 'Volume Provinsi di Wilayah Maluku - Papua Berdasarkan Jenis Olahan',
       x = 'Jenis Olahan', y = 'Volume')
       ).draw()
plt.show()
```

Jenis olahan yang paling banyak mendukung produk olahan ikan di Provinsi Maluku adalah jenis olahan ikan beku.

```python
# Visualisasi data volume produk olahan ikan diwilayah Maluku - Papua berdasarkan jenis perairan
plotnine.options.figure_size = (10,5)
(ggplot(produk_olahan_ikan[produk_olahan_ikan['NamaProvinsi'].isin(['Maluku','Maluku Utara','Papua Barat','Papua'])])

+ aes(x = 'JenisPerairan', y = 'Volume', fill = 'NamaProvinsi')
+ geom_col(position = 'position_dodge')
+ coord_flip()
+ labs(title = 'Volume Provinsi di Wilayah Maluku - Papua Berdasarkan Jenis Perairan',
       x = 'Jenis Perairan', y = 'Volume')
       ).draw()
plt.show()
```

Jenis perairan yang paling banyak mendukung produk olahan ikan di Provinsi Maluku adalah jenis perairan laut.

```python
# Visualisasi data volume produk olahan ikan diwilayah Maluku - Papua berdasarkan tahun
plotnine.options.figure_size = (10,5)
(ggplot(produk_olahan_ikan[produk_olahan_ikan['NamaProvinsi'].isin(['Maluku','Maluku Utara','Papua Barat','Papua'])])

+ aes(x = 'Tahun', y = 'Volume', fill = 'NamaProvinsi')
+ geom_col(position = 'position_dodge')
+ coord_flip()
+ labs(title = 'Volume Provinsi di Wilayah Maluku - Papua Berdasarkan Tahun',
       x = 'Tahun', y = 'Volume')
       ).draw()
plt.show()
```

**Kesimpulan**

Berdasarkan analisis produk olahan ikan pada wilayah Maluku - Papua, Provinsi yang menghasilkan produk olahan ikan paling banyak adalah Provinsi Maluku dengan volume 388.217. Hal ini ditunjukkan dengan grafik volume produk olahan ikan pada tahun 2001-2011 paling banyak terdapat di Provinsi Maluku. ### Insight 6

**Kenapa jenis budidaya kolam memiliki jumlah terbanyak?**

```python
# Visualisasi data jumlah produksi benih berdasarkan budidaya
jumlah_produksi_benih.groupby(['Tahun','Budidaya'])['Jumlah'].sum().unstack().plot(marker='.', cmap='plasma')
plt.title('Jumlah Produksi Benih Tahun 2001-2012',loc='center',pad=30, fontsize=20, color='black')
plt.xlabel('Tahun', fontsize = 15)
plt.ylabel('Jumlah',fontsize = 15)
plt.grid(color = 'darkgray', linestyle = ':', linewidth = 0.5)
plt.ylim(ymin = 0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/100000).astype(int))
plt.legend(loc = 'upper center', bbox_to_anchor = (1.1,1), shadow = True, ncol = 1)
plt.gcf().set_size_inches(12,7)
plt.tight_layout()
plt.show()
```

Berdasarkan grafik diatas dari tahun 2001 sampai dengan 2012 total jumlah budidaya kolam memperoleh hasil terbanyak dibandingkan jenis budidaya lainnya.

```python
# Visualisasi data jumlah produksi benih berdasarkan budidaya
plotnine.options.figure_size = (10,10)
(ggplot(jumlah_produksi_benih[jumlah_produksi_benih['Budidaya'].isin(['budidaya jaring apung','budidaya karamba','budidaya kolam','budidaya laut','budidaya minapadi','budidaya tambak'])])

+ aes(x = 'Benih', y = 'Jumlah', fill = 'Budidaya')
+ geom_col(position = 'position_dodge')
+ coord_flip()
+ labs(title = 'Jumlah Produksi Benih Berdasarkan Budidaya',
       x = 'Benih', y = 'Jumlah')
       ).draw()
plt.show()
```
